```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import pickle
```

```python

data = pd.read_csv('bp.csv')

data
```

```python
X=data[['bp','age']]
X
```

```python
Y=data[['Medication']]
Y
```

```python

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Standardize the features
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Encode the target variable
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

```

```python
import pickle


# Save the StandardScaler object
with open('scaler.pkl', 'wb') as file:
    pickle.dump(sc, file)

# Save the LabelEncoder object
with open('label_encoder.pkl', 'wb') as file:
    pickle.dump(label_encoder, file)
```

```python
X_train
```

```python
X_array=X_test.values

X_test_std
```

```python
X_array
```

```python
# Initialize the SVM classifier
svm = SVC(kernel='linear', C=1.0, random_state=42)

# Train the SVM classifier
svm.fit(X_train_std, y_train_encoded)

# Predict the labels of the test set
y_pred_encoded = svm.predict(X_test_std)

# Calculate accuracy
accuracy = accuracy_score(y_test_encoded, y_pred_encoded)
print("Accuracy:", accuracy)
```

```python
y_pred_encoded
```

```python
predicted_class_decoded = label_encoder.inverse_transform(y_pred_encoded)
predicted_class_decoded
```

```python
custom_input=[[100,80]]

custom_input=np.array(custom_input)

custom_input_std = sc.transform(custom_input)


predicted_class = svm.predict(custom_input_std)

# Decode the predicted class if necessary
predicted_class_decoded = label_encoder.inverse_transform(predicted_class)

arr_str_flat = ', '.join(predicted_class_decoded.flatten())

print("Recommended Medicine:", arr_str_flat)

if predicted_class==0:
    if custom_input[0][1]<=18:
        print("Recommended Dosage Level: 5mg")
    else:
        print("Recommended Dosage Level: 10mg")

elif predicted_class==1:
    if custom_input[0][1]<=18:
        print("Recommended Dosage Level: 2.5mg")
    else:
        print("Recommended Dosage Level: 2.5mg")
else:
    print("The Patient is Normal")
```
